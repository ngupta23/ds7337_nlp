```python
import tensorflow as tf
import pandas as pd
import json
with open("./Downloads/penn-data.json","r") as f:
    data = json.load(f)
```

```python
for i in data:
    print(i)
```

```python
x=[]
y=[]
for i in data:
    x.append(i[0])
    y.append(i[1])
```

```python
import string
from collections import Counter
word_dict = Counter()
table = str.maketrans('', '', ",.")
for i in x:
    words = i.split()
    stripped = [w.translate(table) for w in words]
    for item in stripped:
        word_dict[item]+=1
```

```python
len(word_dict.keys())
```

```python
popular = dict(word_dict.most_common(10000))
```

```python
popular
```

```python
word_map = dict(enumerate(popular.keys(),3))
```

```python
word_map
```

```python
word_map[0]='<pad>'
word_map[1]='<start>'
word_map[2]='<unk>'
```

```python
reverse_word_map = dict([(value, key) for (key, value) in word_map.items()])
```

```python
reverse_word_map['<unk>']
```

```python
cats = Counter()
for i in y:
    for tag in i:
        cats[tag]+=1
```

```python
target_map = dict(enumerate(cats.keys()))
target_map[41]='<pad>'
reverse_target_map = dict([(value, key) for (key, value) in target_map.items()])
```

```python
[reverse_word_map[i] if i in reverse_word_map.keys() else 2 for i in x[0].split()]


for i in x[0].split():
    if i in reverse_word_map.keys():
        reverse_word_map[i]
    else:
        2
```

```python
X_data = []
for sent in x:
    tmp = [reverse_word_map[i] if i in reverse_word_map.keys() else 2 for i in sent.split()]
    X_data.append(tmp)
```

```python
y_data = []
for targets in y:
    tmp = [reverse_target_map[i] for i in targets]
    y_data.append(tmp)
```

```python
from tensorflow.keras.preprocessing import sequence
print(len(X_data))
max_words = 50
X = sequence.pad_sequences(X_data, maxlen=max_words)
y = sequence.pad_sequences(y_data, maxlen=max_words, value = 41)
```

```python
from tensorflow.keras import Sequential, Input
from tensorflow.keras.layers import add
from tensorflow.keras.layers import LSTM, Embedding, Dense, TimeDistributed, Dropout, Bidirectional, Lambda
from tensorflow.keras.layers import SimpleRNN
embedding_vecor_length = 100
vocabulary_size = len(word_map)
n_tags = len(target_map)

model = Sequential()
model.add(Embedding(vocabulary_size, embedding_vecor_length, input_length=max_words))
model.add(Bidirectional(LSTM(units=500, return_sequences=True, recurrent_dropout=0.2, dropout=0.2)))
model.add(TimeDistributed(Dense(n_tags, activation="softmax")))
model.compile(optimizer="adam", loss="sparse_categorical_crossentropy", metrics=["accuracy"])
```
